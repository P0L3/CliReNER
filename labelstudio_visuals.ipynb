## Graph Construction

```python
import json
import networkx as nx
import matplotlib.pyplot as plt

# Load the JSON file
file_path = "DATA/LABEL_STUDIO/project-6-at-2025-07-03-13-30-ef26edce.json" 

#"DATA/LABEL_STUDIO/project-6-at-2025-06-16-10-59-6f9b2621.json"

# "DATA/LABEL_STUDIO/project-6-at-2025-06-06-11-15-c892f102.json"
# "DATA/LABEL_STUDIO/project-6-at-2025-04-22-13-14-67864b63.json"
# "DATA/LABEL_STUDIO/project-6-at-2025-04-07-12-38-0295cb28.json"
# "DATA/LABEL_STUDIO/project-6-at-2025-04-04-08-25-a1f27948.json" 
# "DATA/LABEL_STUDIO/project-6-at-2025-04-02-10-44-5186dca6.json" 
# "DATA/LABEL_STUDIO/project-6-at-2025-04-01-12-54-e72f4afa.json" 
# # "DATA/LABEL_STUDIO/project-6-at-2025-03-31-11-50-c08f75da.json" 
# # "DATA/LABEL_STUDIO/project-6-at-2025-03-27-09-59-771b59ec.json"
with open(file_path, "r", encoding="utf-8") as file:
    data = json.load(file)

# Initialize a directed graph
G = nx.DiGraph()

# Dictionary to store entities and their types
label_to_node = {}  # maps entity text to node id
entity_types = {}   # maps node id to a list of types

def most_common(lst):
    return max(set(lst), key=lst.count)

for entry in data:
    annotations = entry.get("annotations", [])
    for annotation in annotations:
        entities = {e["id"]: e["value"]["text"] for e in annotation.get("result", []) if e["type"] == "labels"}
        entity_types_mapping = {e["id"]: e["value"]["labels"] for e in annotation.get("result", []) if e["type"] == "labels"}
        
        relations = [r for r in annotation.get("result", []) if r["type"] == "relation"]

        # Merge nodes with the same label and collect entity types
        for entity_id, entity_text in entities.items():
            # Merge nodes with the same label and accumulate types
            if entity_text in label_to_node:
                merged_node = label_to_node[entity_text]
                # Add the entity type to the list of types for this node
                existing_types = entity_types.get(merged_node, [])
                new_types = entity_types_mapping.get(entity_id, [])
                entity_types[merged_node] = list(existing_types + new_types)  # Avoid duplicates
            else:
                merged_node = entity_id
                label_to_node[entity_text] = merged_node
                entity_types[merged_node] = entity_types_mapping.get(entity_id, [])  # Initialize type list

            G.add_node(merged_node, label=entity_text, types=most_common(entity_types[merged_node]))  # Add types to node attributes, choose the most frequent type
            # print(entity_types[merged_node])

        # Add relations and include the relation name from the relation_mapping
        for relation in relations:
            from_id = relation["from_id"]
            to_id = relation["to_id"]
            direction = relation["direction"]
            
            relation_type = relation.get("labels", [""])[0]  # Extract first label or empty string
            if relation_type == "":
                relation_type = "0"
            # Map to merged nodes
            from_node = label_to_node.get(entities.get(from_id, ""), from_id)
            to_node = label_to_node.get(entities.get(to_id, ""), to_id)

            relation_name = relation_mapping.get(int(relation_type), "")  # Get relation name from the mapping
            # print(relation_name)
            # Add edge with relation name and relation type ID
            if direction == "right":
                G.add_edge(from_node, to_node, label=relation_name, relation_id=relation_type)
            elif direction == "left":
                G.add_edge(to_node, from_node, label=relation_name, relation_id=relation_type)
            else:
                G.add_edge(from_node, to_node, label=relation_name, relation_id=relation_type)
                G.add_edge(to_node, from_node, label=relation_name, relation_id=relation_type)

# Draw the graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, seed=42)  # Fix seed for consistent layout
node_labels = nx.get_node_attributes(G, 'label')
edge_labels = nx.get_edge_attributes(G, 'label')

# Draw the graph with labels
nx.draw(G, pos, with_labels=True, labels=node_labels, node_size=2500, node_color="lightblue", edge_color="gray", font_size=10)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

plt.title("Merged NER Relations Graph")
plt.show()

# Export to GraphML
graphml_path = file_path.replace(".json", "_info.graphml")
nx.write_graphml(G, graphml_path)
print(f"Graph exported to {graphml_path}")
```

```python
import networkx as nx
import xml.etree.ElementTree as ET

# === Step 1: Load graph and enforce MultiDiGraph ===
path = "./DATA/LABEL_STUDIO/project-6-at-2025-07-03-13-30-ef26edce_info.graphml"
G_orig = nx.read_graphml(path, force_multigraph=True)

# Rebuild as MultiDiGraph to preserve edge data properly
G = nx.MultiDiGraph()
G.add_nodes_from(G_orig.nodes(data=True))
for u, v, data in G_orig.edges(data=True):
    G.add_edge(u, v, **data)

# === Step 2: Find 'is a' edges ===
isa_edges = set()
for u, v, k, data in G.edges(keys=True, data=True):
    if data.get("label") == "is a":
        isa_edges.add((u, v))

# === Step 3: Detect bidirectional 'is a' relationships ===
biisa_pairs = {tuple(sorted((a, b))) for (a, b) in isa_edges if (b, a) in isa_edges}

# === Step 4: Remove bidirectional 'is a' edges ===
for a, b in biisa_pairs:
    keys_to_remove = []
    for u, v, k, data in G.edges(keys=True, data=True):
        if (u, v) in [(a, b), (b, a)] and data.get("label") == "is a":
            keys_to_remove.append((u, v, k))
    for u, v, k in keys_to_remove:
        G.remove_edge(u, v, key=k)

# === Step 5: Add BIISA edges ===
for a, b in biisa_pairs:
    G.add_edge(a, b, label="BIISA", relation_id="-1")

# === Step 6: Save to GraphML ===
output_path = path.replace(".graphml", "_biisa.graphml")
nx.write_graphml(G, output_path)

# === Step 7: Clean XML to remove edge id="..." ===
tree = ET.parse(output_path)
root = tree.getroot()
ns = {"graphml": "http://graphml.graphdrawing.org/xmlns"}

for edge in root.findall(".//graphml:edge", ns):
    if "id" in edge.attrib:
        del edge.attrib["id"]

cleaned_path = output_path.replace(".graphml", "_clean.graphml")
tree.write(cleaned_path, encoding="utf-8", xml_declaration=True)

print(f"Removed {len(biisa_pairs)*2} 'is a' edges and added {len(biisa_pairs)} BIISA edges.")
print(f"Cleaned GraphML saved to: {cleaned_path}")
```

```python
import networkx as nx
# from pyvis.network import Network # Not needed for PyGraphviz
# from IPython.display import IFrame # Not needed for PyGraphviz output
from networkx.algorithms.traversal.breadth_first_search import bfs_edges
import pygraphviz as pgv # Import PyGraphviz
from collections import defaultdict

# === Load graph ===
path = "./DATA/LABEL_STUDIO/project-6-at-2025-07-03-13-30-ef26edce_info_biisa_clean.graphml"

G = nx.read_graphml(path, force_multigraph=True)

# === Step 1: Contract BIISA edges ===
# Find all BIISA edges
biisa_pairs = []
for u, v, data in G.edges(data=True):
    if data.get("label") == "BIISA":
        biisa_pairs.append((u, v))

# Create a mapping for union-find style merging
merged = {}
for u, v in biisa_pairs:
    # Ensure consistent root selection if multiple BIISA edges connect to a node
    root_u = u
    while root_u in merged:
        root_u = merged[root_u]
    
    root_v = v
    while root_v in merged:
        root_v = merged[root_v]

    # Prefer the "smaller" node ID as root for consistency, or any other deterministic rule
    if root_u < root_v:
         merged[root_v] = root_u
    else:
         merged[root_u] = root_v


# Build a mapping from original node to new node
rep_map = {}
for node in G.nodes:
    rep = node
    path_to_root = [rep]
    while rep in merged:
        rep = merged[rep]
        path_to_root.append(rep)
    # Path compression
    for n_in_path in path_to_root:
        rep_map[n_in_path] = rep

# Create a new graph with contracted nodes
G_contracted = nx.MultiDiGraph()

# Add nodes first, ensuring attributes from the representative are preferred
# For node attributes, decide which attributes to keep if nodes are merged.
# Here, we take attributes of the representative node chosen.
for node in G.nodes():
    rep = rep_map[node]
    if rep not in G_contracted:
        # If G.nodes[node] doesn't have data for 'rep', use G.nodes[rep]
        # This assumes 'rep' is an actual node ID from the original graph G
        # which is true by our rep_map construction
        G_contracted.add_node(rep, **G.nodes[rep])
    # else:
        # If you need to merge attributes (e.g., combine labels), do it here.
        # For now, we're just taking the attributes of the first encountered rep.

# Add edges, remapping nodes
for u, v, k, data in G.edges(keys=True, data=True): # Use keys for MultiDiGraph
    u_rep = rep_map[u]
    v_rep = rep_map[v]
    if u_rep != v_rep:  # Skip self-loops from contraction
        # If the original edge was BIISA and got contracted, it's now a self-loop (and skipped)
        # If it was another type of edge between nodes that are now merged, it's also skipped.
        # Only add if it's not a BIISA edge that caused the contraction
        if data.get("label") != "BIISA":
             G_contracted.add_edge(u_rep, v_rep, key=k, **data)
        # If you want to keep BIISA edges that *don't* result in self-loops after contraction
        # (i.e., BIISA between already distinct contracted components), you'd adjust logic.
        # For now, assuming BIISA edges are primarily for contraction.
        # If a BIISA edge remains between u_rep and v_rep, it means they weren't merged by *that* BIISA edge.
        # The current code will add non-BIISA edges. If you want to re-add BIISA edges
        # that didn't cause a self-loop:
        # G_contracted.add_edge(u_rep, v_rep, key=k, **data)


# === Step 2: Find 'EASM' node in contracted graph ===
easm_node = None
for node, data in G_contracted.nodes(data=True):
    if data.get("label") == "EASM": # Assuming 'label' is the attribute storing 'EASM'
        easm_node = node
        break

if easm_node is None:
    print("Warning: No node with label 'EASM' found after contraction.")
    # Fallback: try to find EASM in original graph and map it
    easm_original = None
    for node, data in G.nodes(data=True):
        if data.get("label") == "EASM":
            easm_original = node
            break
    if easm_original:
        easm_node = rep_map.get(easm_original)
        if easm_node:
            print(f"Found EASM original node {easm_original}, mapped to {easm_node} in contracted graph.")
        else:
            raise ValueError("EASM node found in original but not mappable to contracted graph.")
    else:
        raise ValueError("No node with label 'EASM' found in original or contracted graph.")


# === Step 3: Extract neighborhood around EASM ===
N = 2  # Depth of neighborhood
neighbors = set()
if easm_node and easm_node in G_contracted:
    neighbors.add(easm_node)
    for u, v in bfs_edges(G_contracted, easm_node, depth_limit=N):
        neighbors.add(u)
        neighbors.add(v)
    G_sub = G_contracted.subgraph(list(neighbors)).copy() # Ensure it's a list for subgraph
else:
    print(f"EASM node '{easm_node}' not in G_contracted. Visualizing full contracted graph (if small).")
    if G_contracted.number_of_nodes() > 100 : # Arbitrary limit
        print("Contracted graph is too large to visualize fully without EASM focus. Exiting.")
        exit()
    G_sub = G_contracted.copy()


# === Step 4: Visualize with PyGraphviz ===

# Define the set of relation_ids to color
loc = {22, 156, 59, 23, 154, 102, 103, 70, 71, 160, 106, 69, 126, 142}
ce = {3, 10, 25, 27, 125, 28, 111, 60, 78, 29, 139, 79, 76, 62, 30, 77, 85, 133, 104, 105, 53, 68, 58}

# Create a PyGraphviz graph
A = pgv.AGraph(directed=True, strict=False, rankdir='LR', splines='curved') # LR for left-to-right, could be TB for top-bottom
A.graph_attr['label'] = f'EASM Neighborhood (Depth {N}) - Contracted Graph'
A.graph_attr['fontsize'] = '15'
A.node_attr['shape'] = 'ellipse'
A.node_attr['style'] = 'filled'
A.node_attr['fontname'] = 'Helvetica' # Or any print-friendly font
A.edge_attr['fontname'] = 'Helvetica'

# Add nodes to PyGraphviz graph
for node_id, node_data in G_sub.nodes(data=True):
    attrs = {}
    # Use the 'label' attribute from graphml if it exists, otherwise use the node_id
    attrs['label'] = str(node_data.get('label', node_id)) 
    attrs['fillcolor'] = 'lightgreen' # Default node color

    if node_id == easm_node:
        attrs['fillcolor'] = 'orange'
        attrs['shape'] = 'doublecircle' # Make EASM node stand out
        attrs['width'] = '1.0' # Inches
        attrs['height'] = '1.0' # Inches
        attrs['fixedsize'] = 'true'
        attrs['tooltip'] = "EASM (central node)"
        attrs['label'] = f"{attrs['label']}\n(EASM)"
    
    A.add_node(node_id, **attrs)

# Add edges to PyGraphviz graph
# For MultiDiGraph, PyGraphviz handles multiple edges if strict=False.
# If you need to distinguish them by key, you can pass the key to add_edge.
for u, v, edge_data in G_sub.edges(data=True):
    attrs = {}
    edge_label_original = edge_data.get("label", "")
    attrs['label'] = edge_label_original # Default to original label

    # Get relation_id and convert to int if possible
    relation_id_str = edge_data.get("relation_id")
    relation_id = None
    if relation_id_str is not None:
        try:
            relation_id = int(relation_id_str)
        except ValueError:
            print(f"Warning: Could not convert relation_id '{relation_id_str}' to int for edge ({u}-{v})")


    if relation_id is not None and relation_id in loc:
        attrs["color"] = "#DEB887"
        attrs["penwidth"] = "2.0"
    elif edge_data.get("label") == "BIISA": # BIISA edges that weren't contracted out
        attrs["color"] = "red"
        attrs["penwidth"] = "1.0"
    elif "ttribute" in edge_data.get("label", ""): # Ensure label exists before checking 'in'
        attrs["color"] = "grey"
        attrs["penwidth"] = "0.5"
        attrs["label"] = "a" # Override label
        attrs['fontsize'] = '8' # Make 'a' smaller
    elif relation_id is not None and relation_id in ce:
        attrs["color"] = "#800080" # Purple
        attrs["penwidth"] = "3.0"
    else:
        attrs["color"] = "black" # Default edge color
        attrs["penwidth"] = "1.0"
        
    A.add_edge(u, v, **attrs)

prog = "patchwork"
# === Step 5: Layout and Output ===
# Choose a layout engine: dot (hierarchical), neato (spring), fdp (spring), circo (circular), twopi (radial)
# 'neato' or 'fdp' are good for force-directed layouts similar to Pyvis repulsion.
# 'dot' is good for hierarchical structures.
try:
    A.layout(prog=prog) # Try 'dot' first, or 'neato'/'fdp' for less structured graphs
except Exception as e:
    print(f"Failed to layout with 'dot', trying 'neato'. Error: {e}")
    try:
        A.layout(prog='neato')
    except Exception as e2:
        print(f"Failed to layout with 'neato' as well. Error: {e2}")
        # Fallback to no layout, just draw (might be messy)

# Output to print-friendly formats
pdf_file = f"easm_neighborhood_contracted_pygraphviz_{prog}.pdf"
svg_file = f"easm_neighborhood_contracted_pygraphviz_{prog}.svg" # SVG is also great for print
png_file = f"easm_neighborhood_contracted_pygraphviz_{prog}.png" # For quick viewing

try:
    A.draw(pdf_file)
    print(f"PyGraphviz PDF saved to {pdf_file}")
    A.draw(svg_file)
    print(f"PyGraphviz SVG saved to {svg_file}")
    A.draw(png_file)
    print(f"PyGraphviz PNG saved to {png_file}")
except Exception as e:
    print(f"Error during drawing: {e}")
    print("Make sure Graphviz is installed and in your system's PATH.")
    print("On Linux: sudo apt-get install graphviz graphviz-dev")
    print("On macOS: brew install graphviz")
    print("On Windows: download installer from graphviz.org and add bin to PATH.")
```
