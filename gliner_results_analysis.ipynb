## Sampling

Filter the orignal dataset by annotated papers:

```python
# filtered_df = df.loc[df.index[[int(ID) for ID in IDs]]]
filtered_df = df.iloc[[int(ID) for ID in IDs]]
print(len(filtered_df))
```

```python
def list_to_string(row):
    """
    Converts a list of strings into a single space-separated string.
    
    Parameters:
    row (list or str): The input data to be converted.
    
    Returns:
    str: A space-separated string if input is a list, the original string if input is a string,
         or "no_journal" if input is of an unsupported type.
    """
    if isinstance(row, list):
        row = " ".join(row)
        
        
    if isinstance(row, str):
        if "pnas" in row:
            return "PNAS"
        elif "joc" in row:
            return "IJOC"
        elif "eap" in row or "10.1890" in row or "10.2307" in row:
            return "ECOAPP"
        else:
            return row
    else:
        return "no_journal"
```

```python
filtered_df["Journal"] = filtered_df.Journal.apply(list_to_string)
```

```python
JOURNAL_LIST = filtered_df.Journal.unique()
JOURNAL_LIST.sort()
print(JOURNAL_LIST)
```

Create a mapping to original scraped journals (commented parts represent journals with good samples):

```python
J_L = [
    # "Climate Dynamics",
    # "Ecological Applications;ECOAPP",
    # EHS
    # "Energy Policy",
    # "Global Change Biology",
    # "International Journal of Climatology;IJOC",
    # "Journal of Climate",
    # "Journal of Geophysical Research;Journal of Geophysical Research: Atmospheres;Journal of Geophysical Research: Oceans;Journal of Geophysical Research: Space Physics",
    # "Air", "Atmosphere", "Climate", "Earth", "Ecologies", "Energies", "Environments", "Forests", "Fuels", "Hydrology", "Meteorology", "Oceans", "Recycling", "Sustain.", "Water", # MDPI
    # "Nature Climate Change",
    # "Nature Geoscience",
    # NPJ Climate Action
    # NPJ Climate and Atmospheric Science
    # NPJ Ocean Sustainability
    # "PNAS"
]
```

```python
seed(2003)
sampled_rows = []  # List to store the sampled rows

for j in J_L:
    print(f"Processing journal: {j}")
    temp_sample = filtered_df[filtered_df["Journal"].isin(j.split(";"))].sample(60)  # Sample 1 row
    
    for _, row in temp_sample.iterrows():
        print("Title:", row["Title"])
        print("DOI: ", row["DOI"])
        print("Year: ", row["Date"])
        print("backtrace_index:", row["backtrace_index"])
        print()
    
    sampled_rows.append(temp_sample)  # Append the sampled row to the list
    
    print()  # For readability
    print()

# Concatenate all sampled rows into a single DataFrame
new_df = pd.concat(sampled_rows, ignore_index=True)

# If you want to check the final DataFrame
print(new_df)
```

### Copy

Copies sampled papers to a new folder:

```python
import shutil
import os

# List of filenames
file_list = sample_list

# Define the source directory where the files currently are
source_dir = '/home/p0l3/RAD/CWED4ETA/CWED4ETA/CWED4ETA/RESULTS/GLINER_TESTS_FULL'

# Define the destination directory where you want to copy the files
destination_dir = '/home/p0l3/RAD/CWED4ETA/CWED4ETA/CWED4ETA/RESULTS/GLINER_TESTS_SAMPLE'

# Create the destination directory if it doesn't exist
os.makedirs(destination_dir, exist_ok=True)

# Loop through the list and copy each file
for file_name in file_list:
    # Create the full path to the source file
    source_file = os.path.join(source_dir, file_name)
    
    # Create the full path to the destination file
    destination_file = os.path.join(destination_dir, file_name)
    
    # Copy the file
    shutil.copy(source_file, destination_file)
    print(f"Copied {file_name} to {destination_dir}")
```
