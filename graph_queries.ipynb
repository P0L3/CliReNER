## Helper functions

```python
def find_edges_by_node_types(G, head_type_target, tail_type_target, edge_filter = []):
    """
    Find and print all edges in graph G where:
    - the source node has type `head_type_target`
    - the target node has type `tail_type_target`
    """
    matching_edges = []

    for u, v, data in G.edges(data=True):
        head_type = G.nodes[u].get("types", "")
        tail_type = G.nodes[v].get("types", "")
        r_id = data.get("relation_id", "")

        if (head_type == head_type_target or head_type_target == "") and (tail_type == tail_type_target or tail_type_target == "") and int(r_id) not in edge_filter:
            matching_edges.append((u, v, data))

    for u, v, data in matching_edges:
        e1 = G.nodes[u].get("label", "")
        e2 = G.nodes[v].get("label", "")
        r = data.get("label", "")
        r_id = data.get("relation_id", "")
        head_type = G.nodes[u].get("types", "")
        tail_type = G.nodes[v].get("types", "")
        print(f"({head_type}) {e1} - {r} ({r_id}) - {e2} ({tail_type})")
    
    return matching_edges

def find_nodes_by_relation(G, relation_label_target, head_type_target=None, tail_type_target=None, ):
    """
    Find and print all edges in graph G where:
    - the edge has label equal to `relation_label_target`
    - (optional) the source node has type `head_type_target`
    - (optional) the target node has type `tail_type_target`
    - the relation_id is not in `edge_filter`
    """
    matching_edges = []

    for u, v, data in G.edges(data=True):
        r = data.get("label", "")
        r_id = data.get("relation_id", "")
        head_type = G.nodes[u].get("types", "")
        tail_type = G.nodes[v].get("types", "")

        if r != relation_label_target:
            continue

        if r_id == "" or not r_id.isdigit():
            continue

        if head_type_target is not None and head_type != head_type_target:
            continue

        if tail_type_target is not None and tail_type != tail_type_target:
            continue

        matching_edges.append((u, v, data))

    for u, v, data in matching_edges:
        e1 = G.nodes[u].get("label", "")
        e2 = G.nodes[v].get("label", "")
        r = data.get("label", "")
        r_id = data.get("relation_id", "")
        head_type = G.nodes[u].get("types", "")
        tail_type = G.nodes[v].get("types", "")
        print(f"({head_type}) {e1} - {r} ({r_id}) - {e2} ({tail_type})")

    return matching_edges

import json

def load_rel_dict(path):

    with open(path) as f:
        data = f.read()

    rel_dict = {}
    for r in data.split("\n")[1:-1]:
        rel_id, rel = re.split(r"\((\d+)\)",r)[1:]
        rel_dict[int(rel_id)] = rel.strip()

    return rel_dict

import json

def format_relations_from_json(json_string: str, path: str = "", list_entities: bool = False) -> None:
    """
    Parses a JSON string containing sentence annotations and performs one of two actions based on the 'list_entities' flag:
    1. Prints the sentence followed by a human-readable list of its entity relationships.
    2. Lists all entities with their corresponding types.

    Args:
        json_string: A string containing the JSON data in the specified format.
        path: Optional path to a file containing a dictionary for relation labels.
        list_entities: If True, lists entities and their types. Otherwise, prints relationships.
    """
    data = json.loads(json_string)
    sentence = data['data']['sentence']
    annotation_results = data['annotations'][0]['result']

    entity_id_to_text = {}
    entity_id_to_type = {}
    relations = []

    for item in annotation_results:
        if item.get('type') == 'labels':
            entity_id = item['id']
            entity_text = item['value']['text']
            entity_type = item['value']['labels'][0]
            entity_id_to_text[entity_id] = entity_text
            entity_id_to_type[entity_id] = entity_type

        elif item.get('type') == 'relation':
            relations.append(item)

    print(sentence)
    print()
    
    if list_entities:
        for entity_id, entity_text in entity_id_to_text.items():
            entity_type = entity_id_to_type.get(entity_id, 'UNKNOWN_TYPE')
            print(f"{entity_text} - {entity_type}")
    else:

        rel_dict = {}
        if len(path) != 0:
            # Assuming load_rel_dict is a function you have defined elsewhere
            # to load your relation dictionary.
            # For demonstration, it's commented out.
            # rel_dict = load_rel_dict(path)
            pass

        for rel in relations:
            from_id = rel['from_id']
            to_id = rel['to_id']
            try:
                label = rel['labels'][0]
            except (KeyError, IndexError):
                label = "(-2)"

            source_text = entity_id_to_text.get(from_id, 'UNKNOWN_ENTITY')
            target_text = entity_id_to_text.get(to_id, 'UNKNOWN_ENTITY')

            if rel_dict:
                try:
                    print(f"{source_text} --{rel_dict[int(label)]} ({label})--> {target_text}")
                except (ValueError, KeyError):
                    print(f"{source_text} --{label}--> {target_text}")
            else:
                print(f"{source_text} --{label}--> {target_text}")
```
